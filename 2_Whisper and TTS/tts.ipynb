## Text to Speech (TTS) API

```python
import os
os.environ['OPENAI_API_KEY'] = ''
```

```python
# text = 'Viitorul este o carte deschisă, plină de capitole încă nescrise,\
# așteptând strălucirea minților noastre să picteze poveștile lor.'

text = 'In the future, we may be able to travel to other galaxies using wormholes or warp drives.'
```

```python
cost = len(text)/1000 * 0.15  # cost in cents
cost
```

```python
from openai import OpenAI
client = OpenAI()
response = client.audio.speech.create(
     model="tts-1", 
     voice="alloy",  
     input=text
)

# 1. 
response.stream_to_file('tts.mp3')  
# or if you get a DeprecationWarning use this instead:
# 2.
with open('tts-1.mp3', 'wb') as f:
    f.write(response.content)
```
